```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
companies = pd.read_csv('C:\\Users\\avata\\Desktop\\New folder\\ML\\Class 2\\1000_Companies.csv')
```

```python
companies.head()
```

```python
# Separating independent and dependent variable

X = companies.drop(columns = ['Profit']).values
y = companies['Profit'].values
```

```python
print(X)
```

```python
print(y)
```

# Data Processing

```python
labelencoder = LabelEncoder()
```

```python
X[:,3] = labelencoder.fit_transform(X[:, 3]) # which column is categorical
```

```python
print(X)
```

```python
X[:,3]
```

```python
onehot_encoder = OneHotEncoder(categories = 'auto', sparse_output = False)

X = onehot_encoder.fit_transform(X)
```

```python
print(X)
```

# Splitting Data

```python
# tuples
# 0.2 means 20% for test
# random_state = 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100) 
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```
